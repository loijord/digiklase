## Funkcijos samprata ★★☆
1\. Raskite $f(-5)$, kai $f(x-2) = -x^3-x+3$.  
2\. Raskite $f(7)$, kai $f(2x+3) = f(x) - 5x$ ir $f(2) = 4$.  
3\. Funkcija $f(x)$ tenkina lygybę $f(x+2)=2f(x)-3f(x+1)$. Apskaičiuokite $f(3)$, jei $f(1)=11$ ir $f(2)=12$.  
4\. Funkcija $f(x)$ tenkina lygybę $f(x+2)=f(x)+2f(x+1)$. Apskaičiuokite $f(5)$, jei $f(1)=2$ ir $f(2)=3$.  
5\. Raskite $f(5)$, kai $f(x-1) = g(x+1)$ ir $g(x)=x^2+1$.  
6\. Funkcijos $f(x)$ apibrėžimo sritis yra $D_f \in (-4; 11)$. Kokia yra funkcijos $g(x)=2f(x+3)+4$ apibrėžimo sritis?  
7\. Funkcijos $f(x)$ reikšmių sritis yra $E_f \in (-3; 9)$. Kokia yra funkcijos $g(x)=-f(x-1)+5$ reikšmių sritis?  
8\. Funkcijos $f(x)$ reikšmių sritis yra $E_f \in (-8; 5]$. Kokia yra funkcijos $g(x)=|f(x)|-2$ reikšmių sritis?  
9\. Duota funkcija $f(x) = \left\{ \begin{array}{lr} \sqrt[3]{x-1}, & \text{kai}\, x<1 \\ 3x+2a,& \text{kai}\, x\ge 1 \end{array} \right.$. Raskite $a$, kai $f(-7) = f(3)$.  
10\. Duota funkcija $f(x) = \left\{ \begin{array}{lr} tx+8, & \text{kai}\, x<3 \\ x^2-4,& \text{kai}\, x\ge 3 \end{array} \right.$.
* Apskaičiuokite $f(4)$
* Su kuria $t$ reikšme funkcija $f(x)$ yra tolydi visoje apibrėžimo srityje?

11\. Funkcija $f(x)$ yra lyginė ir $f(2)=3$, o funkcija $g(x)$ yra nelyginė ir $g(3)=-2$. Raskite $f(g(-3)) + g(f(-2))$ reikšmę.  
12\. Funkcija $f(x)$ yra lyginė, o funkcija $g(x)$ yra nelyginė. Remdamiesi žemiau pateiktos lentelės duomenimis, apskaičiuokite reiškinio $f(g(-1)) - g(f(-2))$ reikšmę.  
$\begin{array}{|c|c|c|}
\hline x&f(x)&g(x) \\ 
\hline 1 & -1 & -2 \\ 
\hline 2&1&3 \\ 
\hline 3&-1&2 \\
\hline 
\end{array}$
