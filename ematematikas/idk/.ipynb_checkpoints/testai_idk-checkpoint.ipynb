8\. Pirmoje pintinėje yra dvigubai daugiau grybų negu antroje. Kai iš pirmosios pintinės išėmėme $18$ grybų, o į antrą įdėjome $10$ grybų, tai abiejuose pintinėse grybų tapo po lygiai. Kiek grybų buvo pirmoje pintinėje iš pradžių?

11\. Saldainių kilogramas kainuoja $3$ eurais brangiau negu kilogramas sausainių. Už $5$ kilogramus saldainių sumokėta $26$ eurais daugiau negu už $3$ kilogramus sausainių. Kiek eurų kainuoja vienas kilogramas saldainių?
