## Sprendimai
8\. Su tam tikra sveikųjų skaičių pora $(x;\;y)$ skaičių sekos $x$; $xy$; $11x$ ir $y^2$; $x^2$; $x+y+3$ yra aritmetinės proresijos. Nustatykite reiškinio $y^2 - x^2$ reikšmę.  

**Sprendimas.**  

$\begin{cases} xy - x = 11x - xy \\ x^2 - y^2 = x + y + 3 - x^2 \end{cases}$  

Jei $x \neq 0$, tai pirmą lygtį daliname iš $x$:  $\boxed{y - 1 = 11 - y} \Leftrightarrow \boxed{2y = 12} \Leftrightarrow \boxed{y=6}$.  

Tada iš antros lygties $\boxed{x^2 - 36 = x + 9 - x^2} \Leftrightarrow \boxed{2x^2 - x - 45 = 0} \Leftrightarrow \boxed{(x - 5)(2x + 9) = 0}$.  

Sveikoji $x$ reikšmė, tenkinanti lygtį, yra $x = 5$, tada $y^2 - x^2 = 36 - 25 = 11$.  

Jei $x = 0$, tai iš antros lygties $\boxed{- y^2 = y + 3} \Leftrightarrow \boxed{y^2 + y + 3 = 0}$. Ši lygtis neturi sprendinių, nes $D$ neigiamas.   

9\. Aritmetinės progresijos, turinčios $32$ narius, paskutinysis narys lygus $-52{,}5$, o jos skirtumas lygus $-2{,}5$. Apskaičiuokite visų šios progresijos narių sumą.  

**Sprendimas.**  
$x_1 + 31\cdot (-2{,}5) = -52{,}5 \Leftrightarrow x_1 - 77{,}5 = -52{,}5$. Iš čia $x_1 = 25$. Tuomet narių suma $32 \cdot \dfrac{25 + (-52{,}5)}{2} = -440$.

10\. Raskite sumos $a_{61} + a_{62} + a_{63} + \dots + a_{118} + a_{119} + a_{120}$ reikšmę, jei $a_n = -2 + \dfrac{1}{30}n$.  
**Sprendimas.**  
Kai $n$ padidinamas $1$, sekos nario reikšmė padidėja $\dfrac{1}{30}$, vadinasi seka yra aritmetinė progresija su vardikliu $d = \dfrac{1}{30}$.  
$a_{120} = 2$, $a_{61}  = -2 + \dfrac{1}{30}\cdot 61 = \dfrac{1}{30}$. Iš viso sudedame $60$ narių. Atsakymas $60 \cdot \dfrac{\frac{1}{30} + 2}{2} = 30 \cdot \left(\frac{1}{30} + 2 \right) = 61$  

11\. Sudedami natūralieji skaičiai, kuriuos dalijant iš $5$ gaunama liekana $2$. Kiek tokių skaičių buvo sudėta, jei gauta suma lygi $6996$ ?  
**Sprendimas.** 
$a_1 = 2$; $a_n = 2 + 5(n-1)$; $S = n \cdot \dfrac{a_1 + a_n}{2} = n \cdot \dfrac{2 + 2 + 5(n-1)}{2} = n \cdot \dfrac{5n - 1}{2} = \dfrac{5n^2 - n}{2} = 6996$  

$5n^2 - n - 13992 = 0$; $D = 1 - 4\cdot 5 \cdot (-13992) = 279841 = 529^2$; $n_1 = \dfrac{1 + 529}{10} = 53$; $n_1 = \dfrac{1 - 529}{10} = -52{,}8$. Atsakymas: 53 nariai.  

12\. $a_n$ - aritmetinė progresija. $S_{25} = 975$ ir $S_{40} = 2460$. Raskite $a_{33}$.  

**Sprendimas.**  
$S_{25} = \boxed{25 \cdot \dfrac{a_1 + (a_1 + 24d)}{2} = 975} \Leftrightarrow \boxed{2a_1 + 24d = 78}$  
$S_{40} = \boxed{40 \cdot \dfrac{a_1 + (a_1 + 39d)}{2} = 2460} \Leftrightarrow \boxed{2a_1 + 39d = 123}$  

Atėmę pirmą lygtį iš antros, gauname $\boxed{15d = 45} \Leftrightarrow \boxed{d = 3}$. Įstatę $d=3$ į pirmą lygtį, turėsime $\boxed{2a_1 + 72 = 78} \Leftrightarrow \boxed{a_1 = 3}$. Tada $a_{33} = a_1 + 32d = 3 + 32\cdot 3 = 99$.  

13\. Aritmetinės progresijos pirmųjų $n$ narių sumos formulė yra: $S_n = -n(2n+1)$. Raskite šios progresijos dešimtąjį narį. 

**Sprendimas.**  
$a_{10} = S_{11} - S_{10} = -11\cdot(2\cdot 11+1) - (-10\cdot(2\cdot 10+1)) = -253 -(-210)=-43$

14\. Žemiau pateiktos keturių aritmetinių progresijų pirmųjų $n$
 narių sumų formulės. Kuri iš jų yra didėjančiosios aritmetinės progresijos?  
$\begin{array}{l{\quad}l}
\boxed{A}\, S_{n}=\dfrac{(9-n)n}{2}&
\boxed{B}\, S_{n}=-2n^2-n\\
\boxed{C}\, S_{n}=\dfrac{n(3n-5)}{2}&
\boxed{D}\, S_{n}=3n
\end{array}$  
**Sprendimas (žr. taip pat kitos pastabos)**  
Galima išvesti, kad bet kurios aritmetinės progresijos sumos formulė $S_n = An^2 + Bn$. Tuomet $\begin{cases} \boxed{d = 2A} \\ a_1 = A + B \end{cases}$. Aritmetinė progresija didėja, kai $\boxed{d>0} \Leftrightarrow \boxed{A > 0}$. Renkamės variantą $C$, nes tik šiuo atveju $S_n$ yra formos $An^2 + Bn$ ir koefcientas $A$ yra teigiamas.  

15\. Gamykloje gaminamos tam tikros rūšies detalės. Pirmąją dieną buvo pagamintos $4$ detalės, o kiekvieną kitą dieną buvo pagaminta $3$ detalėmis daugiau nei praėjusią dieną. Žinoma, kad per $10$ dienų buvo pagaminta $7\%$ visų planuojamų pagaminti detalių. Po kelių dienų buvo įvykdytas planas? 

Pirmųjų $10$ dienų suma: $4 + 7 + 10 + \dots + 31 = 10\cdot \dfrac{4 + 31}{2} = 175$.  

$\boxed{\begin{array}{rcl} 175 & - & 7\% \\ x & - & 100\% \end{array}} \Leftrightarrow \boxed{7x = 175\cdot 100} \Leftrightarrow \boxed{x=2500}$    

Tegu $n$ - dienų skaičius. Tada:  
$\boxed{n \cdot \dfrac{4 + 4 + (n-1)\cdot 3}{2} = 2500} \Leftrightarrow \boxed{n \cdot (5 + 3n) = 5000} \Leftrightarrow \boxed{3n^2 + 5n - 5000 = 0} \Leftrightarrow n_1 = \dfrac{-5 + \sqrt{5^2 - 4\cdot 3 \cdot (-5000)}}{2\cdot 3} =  \dfrac{-5 + 245}{2\cdot 3} = 40$.  

Kitas sprendinys neigiamas. 

## Pastabos
Kai kuriuose uždaviniuose buvo pateiktos pirmųjų $n$ narių sumos formulės $S_n = f(n)$, kur $f(n)$ yra kvadratinė funkcija.  
**Uždavinys.** Pagal pateiktą pirmųjų $n$ narių sumos formulę $S_n = f(n)$ nustatyti $n$ - tojo nario $a_n$ formulę ir skirtumo $d$ formulę (**šios formulės žinoti nebūtina**).   
**Sprendimas.** Randame $S_n$ išraišką ir pertvarkome gautą reiškinį taip, kad šioje išraiškoje matytųsi kvadratinės funkcijos koeficientai.  
* $S_n = n \left(\dfrac{a_{1} + a_n}{2}\right) = n \left(\dfrac{a_{1} + \left(a_1 + (n-1)d\right))}{2}\right) = 
n \left(\dfrac{2a_{1} + nd - d)}{2}\right) = n\left(n\cdot\dfrac{d}{2} + a_1 - \dfrac{d}{2}\right) = \dfrac{d}{2}\cdot n^2 + \left(a_1 - \dfrac{d}{2}\right)\cdot n = An^2 + Bn$.

Jei $A$ ir $B$ yra žinomi skaičiai, tai turime lygčių sistemą:  
$\begin{cases} \dfrac{d}{2} = A \\ a_1 - \dfrac{d}{2} = B \end{cases} \Leftrightarrow \begin{cases} \boxed{d = 2A} \\ a_1 = A + B \end{cases}$  
Iš čia $a_n = a_1 + d\cdot(n - 1) = A + B + 2A\cdot (n-1) = A + B + 2An - 2A = \boxed{2An + B - A}$

**Pavyzdys.** Rasime $a_n$, kai $S_n = -n(2n+1)$ (žr. 13 uždavinį).  
**Sprendimas.** $S_n = -n(2n+1) = -2n^2 - n$. Tada $A = -2$, $B=-1$. Remdamiesi gautomis išvadomis, turime:  
$\begin{cases} a_n = 2An + B - A = -4n + 1 \\ d = -4 \end{cases}$  
**Patikrinimas.** $S_n = n \cdot \left(\dfrac{a_{1} + a_n}{2}\right) = n \cdot \dfrac{-3 -4n+1}{2} = n \cdot \dfrac{-2 -4n}{2} = n(-2n - 1) = -n(2n+1)$

## Programavimo pavyzdžiai

```python
import numpy as np

def ar_progression(a_1, d, lenght): 
    terms = np.r_[a_1, np.repeat(d, lenght-1)]
    return np.cumsum(terms)

print('Aritmetinė progresija (9 uždavinys):', ar_progression(25, -2.5, 32).tolist())
print('Narių suma:', sum(ar_progression(25, -2.5, 32).tolist()))
print('-'*80)

with np.printoptions(precision=4):
    print('Aritmetinės progresijos posekis (10 uždavinys):', ar_progression(1/30, 1/30, 60))
    print('Narių suma:', np.sum(ar_progression(1/30, 1/30, 60)).round(5))
    print('-'*80)

print('Aritmetinė progresija (11 uždavinys):', ar_progression(2, 5, 53).tolist())
print('Narių suma:', sum(ar_progression(2, 5, 53)))
print('-'*80)

print('Aritmetinės progresijos pirmi 10 narių (12 uždavinys):', ar_progression(3, 3, 10).tolist())
print('S_25:', sum(ar_progression(3, 3, 25)))
print('S_40:', sum(ar_progression(3, 3, 40)))
print('a_33:', ar_progression(3, 3, 33)[-1])
print('-'*80)

print('Aritmetinės progresijos pirmi 10 narių (15 uždavinys):', ar_progression(4, 3, 10).tolist())
print('S_10:', sum(ar_progression(4, 3, 10)))
print('S_40:', sum(ar_progression(4, 3, 40)))
print('-'*80)
```
