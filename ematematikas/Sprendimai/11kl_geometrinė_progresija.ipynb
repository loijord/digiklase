## Sprendimai
1\. Kuri iš duotųjų sekų yra geometrinė progresija?  
$\begin{array}{l{\quad}l}
\boxed{A}\, 5,\,10,\,15&
\boxed{B}\, -9,\,3,\,-1 \\
\boxed{C}\, 3,\,9,\,18 &
\boxed{D}\, 1,\,-7,\,-49
\end{array}$  

**Sprendimas**  
$\boxed{B}$ variantui turime geometrinės progresijos vardiklį $q = \dfrac{3}{-9} = \dfrac{-1}{3}$.

2\. Duoti pirmieji du geometrinės progresijos nariai: $18$; $6$. Kuri iš šių formulių aprašo duotąją seką?  
$\begin{array}{l{\quad}l}
\boxed{A}\, b_n=\dfrac{54}{3^n}&
\boxed{B}\, b_n=30-12n \\
\boxed{C}\, b_n=18\cdot\left(\dfrac{1}{3}\right)^n &
\boxed{D}\, b_n=18-3(n-1)
\end{array}$  
Turime $q = \dfrac{6}{18} = \dfrac{1}{3}$; $b_n = b_1 \cdot q^{n-1} = 18 \cdot \left(\dfrac{1}{3}\right)^{n-1} = 18 \cdot \left(\dfrac{1}{3}\right)^{n} \cdot \left(\dfrac{1}{3}\right)^{-1} = 18 \cdot \left(\dfrac{1}{3}\right)^{n} \cdot 3 = 54 \cdot \left(\dfrac{1}{3}\right)^{n}$.  
Matome, kad šis reiškinys tapatus su variantu $\boxed{A}$  

3\. Seka $b_n$ - geometrinė progresija. Žinoma, kad: $b_5 = 2$ ir $b_9 = \dfrac{2}{81}$. Raskite $b_7$.  

**Sprendimas 1.**  
$b_7 = b_5 \cdot q^2 = \sqrt{b_5 \cdot (b_5 \cdot q^4)} = \sqrt{b_5 \cdot b_9} = \sqrt{2 \cdot \dfrac{2}{81}} = \sqrt{\dfrac{4}{81}} = \dfrac{2}{9}$  
**Sprendimas 2.**  
$q^4 = \dfrac{b_9}{b_5} = \dfrac{\frac{2}{81}}{2} = \dfrac{1}{81} \Rightarrow q = \pm\dfrac{1}{3}$.  
$b_7 = b_5 \cdot q^2 = b_5 \cdot \left(\pm\dfrac{1}{3}\right)^2 = b_5 \cdot \dfrac{1}{9} = \dfrac{2}{9}$.  

4\. Su tam tikra $x$ reikšme skaičių seka $2^x;\;108;\;3^{x+2}$ yra geometrinė progresija. Kam lygus jos vardiklis?  
**Sprendimas.**  
Seka yra geometrinė progresija, kai $\boxed{\dfrac{108}{2^x} = \dfrac{3^{x+2}}{108}} \Leftrightarrow \boxed{\dfrac{108}{2^x} = \dfrac{3^x \cdot 3^2}{108}} \Leftrightarrow \boxed{\dfrac{108}{2^x} = \dfrac{3^x \cdot 9}{108}} \Leftrightarrow \boxed{\dfrac{108}{2^x} = \dfrac{3^x}{12}} \Leftrightarrow \boxed{108 \cdot 12 = 2^x \cdot 3^x} \Leftrightarrow \boxed{1296 = 6^x} \Leftrightarrow \boxed{6^4 = 6^x} \Leftrightarrow \boxed{x=4}$.  
Geometrinės progresijos vardiklis: $q = \dfrac{108}{2^x} = \dfrac{108}{2^4} = \dfrac{4\cdot 27}{4 \cdot 4} = \dfrac{27}{4} = {\large {6}}\dfrac{3}{4}$.  
5\. Duotos keturių geometrinių progresijų bendrojo nario formulės. Kuri iš jų yra mažėjančiosios geometrinės progresijos?  
$\begin{array}{l{\quad}l}
\boxed{A}\, b_n=\dfrac{1}{5}\cdot 4^n &
\boxed{B}\, b_n=-3\cdot \left(\dfrac{1}{2}\right)^n \\
\boxed{C}\, b_n=5^{3-n} &
\boxed{D}\, b_n=4\cdot \left(-\dfrac{2}{3}\right)^n
\end{array}$  
**Sprendimas.** Nagrinėjame kiekvieną iš pateiktų variantų.  
* $\boxed{A}$ variante $b_1 = \dfrac{4}{5}$ ir $q = 4$. Progresijos nariai artėja link $+\infty$.
* $\boxed{B}$ variante $b_1 = -\dfrac{3}{2}$ ir $q = \dfrac{1}{2}$. Progresija nykstamoji, jos nariai artėja link $0$ iš kairės (visi nariai mažesni už $0$).
* $\boxed{C}$ variante $b_1 = 25$, $b_2 = 5$ ir $q = \dfrac{1}{5}$. Progresija nykstamoji, jos nariai artėja link $0$ iš dešinės (visi nariai didesni už $0$).
* $\boxed{D}$ variante $b_1 = -\dfrac{8}{3}$, $q = -\dfrac{2}{3}$. Progresija nykstamoji, jos nariai alteruoja (artėja link $0$ iš abiejų pusių).

Renkamės variantą $\boxed{C}$.  
6\. Geometrinės pirmasis narys lygus $3$. Jos ketvirtojo ir septintojo narių suma yra $4$ kartus didesnė už jos trečiojo ir šeštojo narių sumą. Kam lygus penktasis šios progresijos narys?  
**Sprendimas.**  
$\boxed{b_4 + b_7 = 4(b_3 + b_6)} \Leftrightarrow \boxed{q^3b_1 + q^6b_1 = 4(q^2b_1 + q^5b_1)} \Leftrightarrow \boxed{b_1\cdot q^3 \cdot (1 + q^3) = 4b_1\cdot q^2 \cdot (1 + q^3)} \Leftrightarrow \boxed{q = 4}$.  
$b_5 = b_1\cdot q^4 = 3 \cdot 4^4 = 3 \cdot 256 = \boxed{768}$.  
7\. $b_n$ - geometrinė progresija. Jos pirmasis narys yra $4$, o vardiklis lygus $\dfrac{24}{25}$. Kiek ši geometrinė progresija turi narių, didesnių už $1$?  
**Sprendimas.**  
$\boxed{b_n > 1} \Leftrightarrow \boxed{b_1\cdot q^{n-1} > 1} \Leftrightarrow \boxed{4\cdot \left(\dfrac{24}{25}\right)^{n-1} > 1} \Leftrightarrow \boxed{\left(\dfrac{24}{25}\right)^{n-1} > \dfrac{1}{4}} \Leftrightarrow \boxed{n - 1 < \log_{\frac{24}{25}}\dfrac{1}{4}} \Leftrightarrow \boxed{n < 1 + \log_{\frac{24}{25}}\dfrac{1}{4} \approx 1 + 33{,}96 = 34{,}96}$  
Atsakymas: 34 narius ($b_{34} = 4 \cdot \left(\dfrac{24}{25}\right)^{33} \approx 1{,}04$ ir $b_{35} = 4 \cdot \left(\dfrac{24}{25}\right)^{34} \approx 0{,}998$) 

8\. $b_n = \dfrac{3}{20} \cdot 4^{n-1}$. Kam lygi suma $b_1+b_2+\dots+b_{100}$?  
**Sprendimas.**  
$b_n = b_1 \cdot q^{n-1} = \dfrac{3}{20} \cdot 4^{n-1}$  
Pateikta lygybė yra teisinga su tam tikra $b_1$ reikšme ir kiekvienam $n \in \mathbb{N}$. Iš čia $b_1 = \dfrac{3}{20}$ ir $q = 4$ (Ar galite šias gautas lygybes pagrįsti?).  
Skaičiuojant ieškomą sumą siūlau pakartoti formulės išvedimą:  
$b_1 + b_2 + \dots + b_n = b_1 + b_1q + b_1q^2 + \dots + b_1q^{n-1} = b_1(1 + q + q^2 + \dots + q^{n-1}) = b_1\cdot \dfrac{q^n - 1}{q - 1} = \dfrac{3}{20} \cdot \dfrac{4^n - 1}{3} = \dfrac{1}{20} \cdot (4^n - 1)$.  
$b_1 + b_2 + \dots + b_{100} = \dfrac{4^{100} - 1}{20}$.    

9\. Duota seka: $b_n = 3\cdot 2^n$. Raskite $n$, su kuriuo $S_n = 196602$.  
**Sprendimas.**  
$\boxed{b_1q^{n-1} = 3\cdot 2^n} \Leftrightarrow \boxed{b_1q^{n-1} = 3\cdot 2 \cdot 2^{n-1}} \Leftrightarrow \boxed{b_1q^{n-1} = 6 \cdot 2^{n-1}} \Leftrightarrow \boxed{b_1 = 6\text{ ir }q = 2}$  
Šios geometrinės progresijos sumos formulė $S_n = b_1 \cdot \dfrac{q^n - 1}{q - 1} = 6 \cdot (2^n - 1)$  
$\boxed{6 \cdot (2^n - 1) = 196602} \Leftrightarrow \boxed{2^n - 1 = 32767} \Leftrightarrow \boxed{2^n = 32768} \Leftrightarrow \boxed{n = \log_{2} 32768} \Leftrightarrow \boxed{n = 15}$.  

10\. Kam lygi suma: $216 + 36 + 6 + 1 + \dots$?  
**Sprendimas.**  
$S = b_1 \cdot \dfrac{1}{1-q} = 216 \cdot \dfrac{1}{1-\frac{1}{6}} = 216 \cdot \dfrac{1}{\frac{5}{6}} = 216 \cdot \dfrac{6}{5} = 259{,}2$.  

11\. Raskite nykstamosios geometrinės progresijos teigiamą vardiklį, kai žinoma, kad jos pirmųjų dviejų narių suma lygi $10{,}5$, o visų kitų narių (t.y. pradedant trečiuoju) suma lygi $0{,}3$.  

**Sprendimas.**  
$\begin{cases} b_1 \cdot \dfrac{1}{1-q} = 10{,}8 \\ b_1 + b_1q = 10{,}5 \end{cases} 
\Leftrightarrow \begin{cases} b_1 = 10{,}8 (1 - q) \\ b_1 (1 + q) = 10{,}5\end{cases}$

Iš antros lygties $\boxed{10{,}8 (1 - q)(1 + q) = 10{,}5} \Leftrightarrow \boxed{108(1 - q^2) = 105} \Leftrightarrow \boxed{1 - q^2 = \dfrac{105}{108}} \Leftrightarrow \boxed{q^2 = \dfrac{3}{108} = \dfrac{1}{36}} \Leftrightarrow \boxed{q = \dfrac{1}{6}\,\text{jei }q>0}$

12\. Geometrinės progresijos pirmųjų $n$ narių sumos formulė yra: $S_n = \dfrac{1-3^n}{9}$. Kam lygus dešimtasis šios progresijos narys? 

**Sprendimas 1.** (nesurandant sekos nario formulės)  
$b_{10} = S_{10} - S_{9} = \dfrac{1-3^{10}}{9} - \dfrac{1-3^{9}}{9} = \dfrac{(1-3^{10}) - (1-3^{9})}{9} = \dfrac{3^{9} - 3^{10}}{9} = 3^7 - 3^8 = 3^7\cdot (1 - 3) = -2 \cdot 3^7 = -2 \cdot 2187 = \boxed{-4374}$  

**Sprendimas 2.** (surandant $b_1$ ir $q$)  
$b_1 = S_1 = \dfrac{1-3^1}{9} = -\dfrac{2}{9}$.  
$b_2 = S_2 - S_1 = \dfrac{1-3^2}{9} = -\left(-\dfrac{2}{9}\right) =  -\dfrac{8}{9} + \dfrac{2}{9} = -\dfrac{6}{9}$.  
$q = \dfrac{b_2}{b_1} = 3$.  
$b_{10} = b_1 \cdot q^9 = -\dfrac{2}{9} \cdot 3^9 = -2 \cdot 3^7 = \boxed{-4374}$.  

13\. Raskite lygties $x^{-2}+x^{-4}+x^{-6}+\dots = \dfrac{1}{8}$ sprendinį, jei $x>1$.  

**Sprendimas**  
$x^{-2}$, $x^{-4}$, $x^{-6}$, $\dots$ sudaro nykstamąją geometrinę progresiją, kurios $b_1 = x^{-2}$ ir $q = x^{-2}$ (čia $|q|<1$). Turime:  
$x^{-2}+x^{-4}+x^{-6}+\dots = b_1 \cdot \dfrac{1}{1-q} = x^{-2} \cdot \dfrac{1}{1-x^{-2}} = x^{-2} \cdot \dfrac{x^2}{x^2-1} = \dfrac{1}{x^2 - 1}$.  
Iš čia $\boxed{\dfrac{1}{x^2 - 1} = \dfrac{1}{8}} \Leftrightarrow \boxed{x^2 - 1 = 8} \Leftrightarrow \boxed{x^2 = 9} \Leftrightarrow \boxed{x = 3,\,\text{kai}\,x>1}$  

14\. Didėjančiosios geometrinės progresijos, kurią sudaro $8$ nariai, pirmasis narys lygus $4$, o jos paskutinių keturių ir pirmų`jų keturių narių sumų santykis lygus $625$. Raskite šios progresijos narių sumą.  
**Sprendimas.**  
$\dfrac{b_5 + b_6 + b_7 + b_8}{b_1 + b_2 + b_3 + b_4} = \dfrac{b_1q^4 + b_1q^5 + b_1q^6 + b_1q^7}{b_1 + b_1q + b_1q^2 + b_1q^3} = \dfrac{q^4(1 + q + q^2 + q^3)}{1 + q + q^2 + q^3} = q^4 = 625$.  Su didėjančios progresijos vardikliu $q > 1$ turime $q = 5$.  
Tuomet $S = b_1 \cdot \dfrac{q^8 - 1}{q - 1} = 4 \cdot \dfrac{5^8 - 1}{5 - 1} = 5^8 - 1 = 390624$.  

## Programavimo pavyzdžiai

```python
import numpy as np

def geom_progression(b_1, q, lenght): 
    terms = np.r_[b_1, np.repeat(q, lenght-1)]
    return np.cumprod(terms)

B = geom_progression(3, 4, 7)
print('Geometrinės progresijos pirmi 7 nariai (6 uždavinys):', B.tolist())
print('b_3 + b_6 =', B[2] + B[5])
print('b_4 + b_7 =', B[3] + B[6])
print('b_5 =', B[4])
print('-'*80)

with np.printoptions(precision=3):
    print('Geometrinės progresijos pirmi 35 nariai (7 uždavinys):', geom_progression(4, 24/25, 35))
    print('-'*80)

B = geom_progression(6, 2, 15)
print('Geometrinės progresijos pirmi 15 narių (9 uždavinys):', B.tolist())   
print('Narių suma:', sum(B))
print('-'*80)

with np.printoptions(precision=3, suppress=True):
    B = geom_progression(216, 1/6, 10)
    print('Geometrinės progresijos pirmi 10 narių (10 uždavinys):', B)   
    print('Narių suma:', sum(B))
    print('-'*80)

with np.printoptions(precision=3, suppress=True):
    B = geom_progression(9, 1/6, 10)
    print('Geometrinės progresijos pirmi 10 narių (11 uždavinys):', B)   
    print('Narių suma:', sum(B))
    print('-'*80)

with np.printoptions(precision=5, suppress=True):
    B = geom_progression(1/9, 1/9, 7)
    print('Geometrinės progresijos pirmi 7 nariai (13 uždavinys):', B)   
    print('Narių suma:', sum(B))
    print('-'*80)

B = geom_progression(4, 5, 8)
print('Geometrinė progresija (14 uždavinys):', B.tolist())   
print('Narių suma:', sum(B))
print('-'*80)
```
