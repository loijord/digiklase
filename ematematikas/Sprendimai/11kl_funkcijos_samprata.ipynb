## Funkcijos samprata ★★☆

6\. Funkcijos $f(x)$ apibrėžimo sritis $D_f = (-4; 11)$. Kokia yra funkcijos $g(x)=2f(x+3)+4$ apibrėžimo sritis?    
**Sprendimas.**  
$\boxed{x+3 \in (-4; 11)} \Rightarrow \boxed{x \in (-4-3; 11-3) = (-7; 8)}$. $D_g = (-7; 8)$ (nes $f(x+3)$ apibrėžimo sritis yra tokia pati, kaip ir $2f(x+3)+4$).  

7\. Funkcijos $f(x)$ reikšmių sritis yra $E_f \in (-3; 9)$. Kokia yra funkcijos $g(x)=-f(x-1)+5$ reikšmių sritis?  
**Sprendimas.**  
$E_g = (-9 + 5; -(-3) + 5) = (-4; 8)$ (nes $f(x)$ reikšmių sritis lygi $f(x-1)$ reikšmių sričiai)

8\. Funkcijos $f(x)$ reikšmių sritis yra $E_f = (-8; 5]$. Kokia yra funkcijos $g(x)=|f(x)|-2$ reikšmių sritis?  
**Sprendimas.**  
Kiekvienai reikšmei $f(?) = a$ iš reikšmių srities turėsime $|f(?)|-2 = |a| - 2 = \left \{\begin{array}{lr} a - 2, \quad & a \ge 0 \\ -a - 2, \quad & a<0\end{array} \right.$  
Pirmu atveju $\boxed{a \in [0; 5]} \Rightarrow \boxed{a \in [-2; 3]}$.  
Antru atveju $\boxed{a \in (-8; 0)} \Rightarrow \boxed{-a - 2 \in (-0-2; -(-8)-2) = (-2; 6)}$.  
Galiausiai $E_g = [-2; 3] \bigcup (-2; 6) = [-2; 6]$.

### Pastabos
**Apibrėžimas.** Funkcija laikoma **didėjančia**, jei su visais $a, b \in \mathbb{R}$ yra teisinga: $\boxed{a < b} \rightarrow \boxed{f(a) < f(b)}$. Jei su visais $a, b \in \mathbb{R}$ yra $\boxed{a < b} \rightarrow \boxed{f(a) > f(b)}$, funkcija laikoma **mažėjančia**.  

**Pagalbinis teiginys.** Jei $x \in (a, b)$ ir $f$ yra monotoniška funkcija (didėjanti arba mažėjanti), tai $f(x) \in (f(a); f(b))$. Intervalo kraštai apsikeičia vietomis, kai $f$ yra mažėjanti funkcija. Tai pat gali būti ir uždari.

**Pavyzdys 1.** Jei $f(x) = 2x - 3$, tai $f$ yra didėjanti funkcija. Jei $x \in [-1; 1]$, tai $2x - 3 \in [-1\cdot 2 - 3; 1\cdot 2 - 3] = [-5; -1]$  
**Pavyzdys 2.** Jei $f(x) = -2x + 1$, tai $f$ yra mažėjanti funkcija. Jei $x \in [-1; 1]$, tai $-2x + 1 \in [1\cdot (-2) + 1; -1\cdot (-2) + 1] = [-1; 3]$ (intervalo kraštai susikeitė vietomis)  
**Pavyzdys 3.** Jei $f(x) = x^3 - 5x$, tai $f$ nėra monotoniška funkcija intervale $[-2; 2]$. Turime $f(-2) = 2$ ir $f(2) = -2$, bet $f(-1) = 4$ ir $f(1) = -4$ nepriklauso intervalui $[-2; 2]$.
