1\. Rask skaičių $224$ ir $288$ didžiausią bendrą daliklį  

### Sprendimas I*
* $\dfrac{224}{288} = \dfrac{112}{144} = \dfrac{56}{72} = \dfrac{28}{36} = \dfrac{14}{18} = \dfrac{7}{9}$
* $224:7 = 32$; $288: 9 = 32$
* Vardiklis ir skaitiklis sumažėjo 32 kartus. Vadinasi, $DBD(224, 288) = 32$
