3\. Apskritimo styga, kurios ilgis $3{,}6\,\text{cm}$, nutolusi nuo apskritimo centro atstumu, lygiu $2{,}4\,\text{cm}$. Apskaičiuokite šio apskritimo spindulio ilgį.  
4\. Iš apskritimo taško nubrėžtos dvi tarpusavyje statmenos stygos. Atkarpa, jungianti stygų vidurio taškus lygi $5$. Raskite apskritimo spindulį.  
5\. Vieno apskritimo ilgis yra $5{,}5\pi\,\text{cm}$, o kito apskritimo ribojamas plotas lygus $121\pi\,\text{cm}^2$. Kiek kartų skiriasi šių apskritimų spindulių ilgiai?

9\. Per tašką $A$ nubrėžta apskritimo liestinė, liečianti jį taške $B$. Žinoma, kad $AB=24$ ir $BO=7$. Apskaičiuokite atkarpos $AC$ ilgį.  
<img src="../images/9_12_9.png" width="180"/>  
10\. Per taškus $A$ ir $B$ nubrėžta apskritimo liestinė, liečianti jį taške $C$. Taško $A$ nuotolis nuo apskritimo centro lygus $15$. Žinoma, kad $AC=12$ ir $BC=6{,}75$. Apskaičiuokite taško $B$ nuotolį iki apskritimo centro.  
11\. Iš taško $B$ išvesta apskritimo liestinė, liečianti jį taške $A$. Atkarpos $AB$ ilgis lygus $24$. Ant apskritimo lanko pažymėtas taškas $C$ taip, jog $AC$ - apskritimo skersmuo, kurio ilgis $7$. Sujungus taškus $B$ ir $C$ gauta atkarpa kerta apskritimo lanką taške $D$. Apskaičiuokite stygos $CD$ ilgį.  
12\. Du apskritimai, liečiasi iš išorės. Iš mažesniojo apskritimo, kurio spindulio ilgis lygus $2$, centro $O$ išvesta liestinė didesniajam apskritimui ir liečia jį taške $A$. Žinoma, kad $OA=10$. Apskaičiuokite atstumą tarp apskritimų centrų.
