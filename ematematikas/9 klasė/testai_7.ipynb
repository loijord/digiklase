## Skaičių sekos ★☆☆
1\. Nurodykite devintąjį sekos narį, kai turima seka yra $2,\,7,\,12,\,\dots$

$\begin{array}{l{\quad}l{\quad}l{\quad}l}
\boxed{A}\, 35&
\boxed{B}\, 46&
\boxed{C}\, 53&
\boxed{D}\, 59
\end{array}$  
3\. Kuris iš žemiau pateiktų skaičių priklauso sekai $2,\,9,\,16,\,\dots$ ?  
$\begin{array}{l{\quad}l{\quad}l{\quad}l}
\boxed{A}\, 42&
\boxed{B}\, 43&
\boxed{C}\, 44&
\boxed{D}\, 45
\end{array}$  
4\. Kuriai iš žemiau pateiktų sekų, aprašytų bendrojo nario formulėmis, priklauso skaičius $48$ ?  
$\begin{array}{l{\quad}l{\quad}l{\quad}l}
\boxed{A}\, c_n = 2n - 1&
\boxed{B}\, c_n = 11n + 1&
\boxed{C}\, c_n = 3n - 5&
\boxed{D}\, c_n = 13n - 4
\end{array}$  
5\. Seka aprašoma formule $c_n = 3n - 11$. Raskite šimtąjį šios sekos narį.  
6\. Seka aprašoma formule $c_n = 5n + 9$. Raskite šios sekos nario, lygaus $244$, numerį.  
7\. Seka aprašoma formule $c_n = 3n + 3$. Raskite pirmųjų penkių šios sekos narių sumą.  
8\. Seka aprašoma formule $c_n = n^2 - 7n - 1$. Raskite penktojo ir septintojo šios sekos narių sumą.  
9\. Seka aprašoma formule $c_n = 2n^2 - 23n - 5$. Raskite šios sekos nario, lygaus $65$, numerį.  
10\. seka yra $3,\,6,\,11,\,18,\,\dots$. Kuri iš žemiau pateiktų formulių yra šios sekos bendrojo nario formulė?  
$\begin{array}{l{\quad}l{\quad}l{\quad}l}
\boxed{A}\, c_n = n^2 - 1&
\boxed{B}\, c_n = 3n + 1&
\boxed{C}\, c_n = n^2 + 1&
\boxed{D}\, c_n = 4n - 1
\end{array}$  
11\. Skaičių seka aprašoma rekurentine formule $c_{n+1} = 3c_n + 5$. Jeigu pirmasis sekos narys $c_1=3$, kam yra lygus trečiasis šios sekos narys?  
12\. Skaičių seka aprašoma rekurentine formule $c_{n+1} = 5 - 3c_n$. Jeigu pirmasis sekos narys $c_1=2$, kam yra lygus penktasis šios sekos narys?  
13\. Skaičių seka aprašoma rekurentine formule $c_{n+1} = 2c_n + 3$. Jeigu ketvirtasis sekos narys $c_4=61$, kam yra lygus pirmasis šios sekos narys?  
14\. Skaičių seka aprašoma rekurentine formule $c_{n+1} = 5c_n-55$. Jeigu septintasis sekos narys $c_7=21$, kam yra lygus devintasis šios sekos narys?  
15\. Kiekvienas sekos $a,\,b,\,c,\,12,\,19,\,31,\,50,\,\dots$ narys, pradedant trečiuoju, lygus dviejų prieš jį einančių narių sumai. Kam lygus skaičius $a$?
