2\. Naudodamiesi brėžiniu, nustatykite pavaizduoto stačiojo trikampio kampo $B$ kosinusą.  
<img src="../images/9_11_2.png" width="240"/>  
3\. Naudodamiesi brėžiniu, nustatykite pavaizduoto stačiojo trikampio kampo $A$ tangentą.  
<img src="../images/9_11_3.png" width="240"/>  
4\. Raskite reiškinio $\sin^{2}45^\circ + \sin 30^\circ + \cos^{2}45^\circ + \cos 30^\circ \cdot \text{tg}\,30^\circ$ reikšmę.  
5\. Raskite reiškinio $\sqrt{2} \cdot \sin \dfrac{\pi}{4} + \cos \dfrac{\pi}{6} - \text{tg} \dfrac{\pi}{4}$ reikšmę.    
6\. Suprastinkite reiškinį $1 + (\cos\alpha + \sin\alpha)(\cos\alpha - \sin\alpha)$  
7\. Suprastinkite reiškinį $\sin\alpha \cdot \text{tg}\,\alpha + \cos\alpha$.  
8\. Raskite $\cos \alpha$ reikšmę, kai $\sin\alpha = 0{,}8$.  
9\. Raskite reiškinio $1 + \text{tg}^2\,\alpha$ reikšmę, kai $\sin\alpha = \dfrac{20}{29}$.  
10\. Apskaičiuokite trikampio $\angle ABC$ didumą laipsniais vienetų tikslumu.  
<img src="../images/9_11_10.png" width="240"/>  
11\. Stačiojo trikampio statinių ilgiai $3$ ir $4$. Į šio trikampio įžambinę išvesta pusiaukraštinė. Kokį kampą ji sudaro su trumpesniuoju statiniu? Atsakymą pateikti laipsnio tikslumu.  
12\. Lygiašonės trapecijos trumpesniojo pagrindo ilgis $4\,\text{cm}$, šoninės kraštinės ilgis $8\,\text{cm}$, kuri su ilgesniuoju pagrindu sudaro $50^\circ$ kampą. Apskaičiuokite trapecijos plotą. Atsakymą pateikite vienetų tikslumu.
