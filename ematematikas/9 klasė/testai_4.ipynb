## Tiesinė funkcija ★★☆
1\. Kuris iš žemiau pateiktų taškų priklauso funkcijos $f(x) = -5x + 6$ grafikui?  
$\begin{array}{l{\quad}l{\quad}l{\quad}l}
\boxed{A}\, (-1; 1) &
\boxed{B}\, (-2; 16) &
\boxed{C}\, (-4; 2) &
\boxed{D}\, (1; -1) &
\end{array}$  
2\. Raskite taško, kuriame funkcija $f(x) = 2x + 10$ kerta $Ox$ ašį, koordinates.  
3\. Nurodykite koordinačių plokštumos taško $x$, kuriame tiesinė funkcija $f(x)=-5x+3$ kerta $Oy$ ašį, reikšmę.  
4\. Funkcijos $f(x) = kx - 1$ grafikas eina per tašką, kurio koordinatės yra $(4; 11)$. Raskite koeficiento $k$ reikšmę.  
5\. Funkcijos $f(x) = kx + b$ grafikas eina per taškus, kurių koordinatės yra $(0;  6)$ ir $(3;  15)$. Raskite koeficiento $k$ reikšmę.  
6\. Taškai $A(2; 6)$ ir $B(-3; 1)$ priklauso tiesei $y = kx + b$. Apskaičiuokite reiškinio $3k+5b=?$ reikšmę.  
7\. Su kuria $x$ reikšme funkcija $g(x) = \dfrac{x + 8}{9}$ lygi $2$?  
8\. Tiesė $y = -4x - 3$ yra lygiagreti kitai tiesei $y = kx + b$. Raskite koeficiento $k$ reikšmę.  
9\. Kuri iš žemiau nurodytų tiesių yra statmena tiesei $y = \dfrac{1}{2}x - 2$?  
$\begin{array}{l{\quad}l}
\boxed{A}\, y = 2x + 3&
\boxed{B}\, y = -\dfrac{1}{2}x + 1\\
\boxed{C}\, y = \dfrac{1}{2}x - 4&
\boxed{D}\, y = -2x + 5
\end{array}$  
10\. Kuris teiginys apie funkcijos $f(x) = -8$ grafiką yra teisingas?  
$\begin{array}{l}
\boxed{A}\, \text{Grafikas yra tiesė, lygiagreti su $Ox$ ašimi ir kertanti $Oy$ ašį taške $(0; -8)$} \\
\boxed{B}\, \text{Grafikas yra tiesė, lygiagreti su $Oy$ ašimi ir kertanti $Ox$ ašį taške $(-8; 0)$} \\
\boxed{C}\, y = \text{Grafikas yra tiesė, nelygiagreti nė su viena koordinačių plokštumos ašimi} \\
\boxed{D}\, \text{Grafikas nėra tiesė}
\end{array}$  
11\. Raskite tiesių $y=4x+5$ ir $y=8x+9$ susikirtimo taško ordinatę.
