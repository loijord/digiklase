## Bandomasis PUPP Nr. M | 2025
* Patikrinimą sudaro 30 uždavinių.
* Patikrinimo taškų suma yra 50 taškų.
* Patikrinimo trukmė – 150 min.
* Šaltinis: [Ematematikas](https://www.ematematikas.lt/testai/10)
* Pastabos.
    * *Testas išleistas prieš oficialų PUPP bandomąjį, todėl įvairios temos pagal taškus gali pasiskirstyti neproporcingai*
    * *Nėra pateikiama, kuriuose uždaviniuose reikia sprendimų. 2025 metų standartas būtų 5 - 6 tokie uždaviniai.*

1\. Mokykloje mokosi $128$ mergaitės, kurios sudaro $80\%$ berniukų skaičiaus. Kiek moksleivių mokosi mokykloje? (**1 tšk.**)  

2\. Suprastink reiškinį $\dfrac{x^2-4x}{x^2-16}$ (**1 tšk.**)  

$\begin{array}{l{\quad}l{\quad}l{\quad}l}
\boxed{A}\, \dfrac{x}{x+4} &
\boxed{B}\, \dfrac{x}{4} & 
\boxed{C}\, \dfrac{x}{x-4} &
\boxed{D}\, \dfrac{1-x}{4}
\end{array}$  

3\. Duota skaičių seka $4$, $7$, $10$, $13$, $\dots$. Kuris iš žemiau pateiktų skaičių nepriklauso šiai sekai? (**1 tšk.**)  
$\begin{array}{l{\quad}l{\quad}l{\quad}l}
\boxed{A}\, 31 &
\boxed{B}\, 43 & 
\boxed{C}\, 53 &
\boxed{D}\, 67
\end{array}$ 

4\. Apskaičiuokite lygties $(x+1)(x-7)=5$ diskriminanto reikšmę. (**1 tšk.**)  

5\. Apskaičiuokite reiškinio $\dfrac{5x-4y}{2x+3y}$ skaitinę reikšmę, kai $0{,}2x = 2y$. (**1 tšk.**)  

6\. Kiekvienas sekos $x$, $y$, $z$, $-1$, $8$, $-9$, $17$, $\dots$ narys, pradedant trečiuoju, lygus dviejų prieš jį einančių narių skirtumui. Kam lygus skaičius $x$? (**1 tšk.**)  

7\. Su kuria parametro $a$ reikšme lygtis $2x^2+2ax+3a=0$ turi vieną sprendinį? (**1 tšk.**)  

8\. Nurodykite nelygybės $10x-x^2>0$ didžiausią sveikąjį sprendinį (**1 tšk.**)  

9\. Suprastinkite reiškinį $(\sin x + \cos x)^2 - 2\text{tg}x\cdot \cos ^2{x}$  (**1 tšk.**)  
$\begin{array}{l{\quad}l{\quad}l{\quad}l}
\boxed{A}\, \sin x &
\boxed{B}\, \cos x & 
\boxed{C}\, 0 &
\boxed{D}\, 1
\end{array}$  

10\. Koordinačių plokštumoje turime tris taškus: $A(1; 5)$, $B(1; -1)$ ir $C(6;-1)$. Sujungus taškus gauname trikampį. Koks jis yra? (**1 tšk.**)  
$\begin{array}{l{\quad}l}
\boxed{A}\, \text{Lygiašonis} &
\boxed{B}\, \text{Lygiakraštis} \\
\boxed{C}\, \text{Statusis} &
\boxed{D}\, \text{Neįmanoma nustatyti}
\end{array}$  

11\. Nurodykite koordinačių plokštumos taško koordinates, kuriame tiesė $y=-5x+15$ kerta $Ox$ ašį (**1 tšk.**)  

12\. Išspręskite nelygybę $x^2 +6x + 5 \geq 0$ ir nurodykite jos sprendinių aibę  (**1 tšk.**)  
$\begin{array}{l}
\boxed{A}\, x\in[-5;-1] \\
\boxed{B}\, x\in(-\infty;-5]\cup[-1;+\infty) \\
\boxed{C}\, x\in[-1;+\infty) \\
\boxed{D}\, x\in(-\infty;+\infty)
\end{array}$  

13\. Turime keturis koordinačių plokštumos taškus: $A(7;\, 28)$, $B(-3;\, -23)$, $C(-4;\, -27)$, $D(0{,}2;\, -6)$. Keli iš jų priklauso tiesės $y=5x-7$ grafikui?  (**1 tšk.**)  

14\. Butelyje yra $1800$ gramų saldaus gėrimo. $5$ procentai šio gėrimo masės sudaro cukrus.  
* Kiek gramų gryno cukraus yra šiame butelyje?  (**1 tšk.**)  
* Kiek gramų vandens reikėtų įpilti į šį gėrimą, kad cukrus sudarytų $4$ procentus jo masės? (**2 tšk.**)

16\. Išspręskite lygčių sistemą $\begin{cases} 12+y=6x \\ x^2=y+3 \end{cases}$ ir apskaičiuokite reiškinio $x+y$ reikšmę (**2 tšk.**)    
17\. Suraskite parabolės $y=ax^2$ koeficiento $a$ reikšmę, kai parabolė eina per tašką $\left(\dfrac{1}{4};2\right)$ (**1 tšk.**)  

18\. Upės tėkmės greitis $2{,}5 \text{km/h}$. Motorinė valtis $35\,\text{km}$ atstumą šia upe prieš srovę nuplaukia per $2\,\text{h}\,20\,\text{min}$. Per kiek valandų ši motorinė valtis nuplauktų $35\,\text{km}$ atstumą, plaukdama šia upe pasroviui?   (**3 tšk.**)  

19\. Nurodykite trupmeninės lygties $\dfrac{x+2}{x^2-4}=0$ sprendinių skaičių (**1 tšk.**)  

20\. Suraskite funkcijos $f(x)=x^2-4x+6$ reikšmių sritį (**1 tšk.**)  
$\begin{array}{l{quad}l{quad}l{quad}l}
\boxed{A}\, x\in R &
\boxed{B}\, [2;+\infty) &
\boxed{C}\, x\in R\backslash\{2\} &
\boxed{D}\, [6;+\infty)
\end{array}$  

21\. Lygiakraščio trikampio plotas lygus $9\sqrt{3}$. Raskite jo perimetrą (**1 tšk.**)  
$\begin{array}{l{quad}l{quad}l{quad}l}
\boxed{A}\, 6\sqrt{3} &
\boxed{B}\, 12 &
\boxed{C}\, 15\sqrt{3} &
\boxed{D}\, 18
\end{array}$

$\begin{array}{|c|c|} \hline    \text{Pažymys} & 5 & 6 & 7 & 8 & 9 & 10 \\\hline   \text{Dažnis} & 2 & 4 & 6 & 8 & 3 & 2 \\\hline \end{array}$  

* Apskaičiuokite šios klasės pažymių vidurkį (**1 tšk.**)  
* Raskite šios klasės moksleivių pažymių medianą (**1 tšk.**)
