### Uždavinys 1.
Duota funkcija $y = f(x) = x^2 + 4x$. Kuris iš nurodytų taškų priklauso šios funkcijos grafikui? (**1 tšk.**)  
$\begin{array}{l{\quad}l{\quad}l{\quad}l}
\boxed{A}\, (1;\,5) &
\boxed{B}\, (2;\,16) & 
\boxed{C}\, (4;\,24) &
\boxed{D}\, (3;\,18)
\end{array}$ 

**Pagalbinis teiginys**. Taškas $({\color{blue}{a}}; {\color{green}{b}})$ priklauso pateiktos funkcijos $y = f(x)$ grafikui tada ir tik tada, kai pakeitus $\begin{cases} {\color{blue}{x}}\to a \\ {\color{green}{y}}\to b \end{cases}$ lygybė tampa teisinga.
      
**Sprendimas.**
* $5 = 1^2 + 4\cdot 1$, todėl $(1;\,5)$ priklauso funkcijos $y = x^2 + 4x$ grafikui.
 
**Paaiškinimas 1A**
Kaip patikrinti, ar taškas $(1;\,5)$ priklauso funkcijos $y = x^2 + 4x$ grafikui?
* Funkcijos formulė: $\boxed{{\color{green}{y}} = {\color{blue}{x}}^2 + 4{\color{blue}{x}}}$
* Pagal **pagalbinio teiginio** sąlygą formulėje pakeisti: $\begin{cases} {\color{blue}{x}}\to 1 \\ {\color{green}{y}}\to 5\end{cases}$
* Pakeista funkcijos formulė:
    * $\boxed{5} = \boxed{1}^2 + 4\boxed{1}$
    * $5 = 1^2 + 4\cdot 1$
* Skaičiavimai. $1^2 + 4\cdot 1 = 1 + 4 = 5$, todėl gauta lygybė $5 = 1^2 + 4\cdot 1$ $\color{green}{\text{teisinga}}$
* Išvada pagal **pagalbinį teiginį**. Taškas $(1;\,5)$ $\color{green}{\text{priklauso}}$ funkcijos $y = x^2 + 4x$ grafikui.

**Paaiškinimas 1B**
Kaip patikrinti, ar taškas $(2;\,16)$ priklauso funkcijos $y = x^2 + 4x$ grafikui?
* Pagal **pagalbinį teiginį** formulėje pakeisti: $\begin{cases} {\color{blue}{x}}\to 2 \\ {\color{green}{y}}\to 16\end{cases}$
* Pakeista funkcijos formulė:
    * $\boxed{16} = \boxed{2}^2 + 4\boxed{2}$
    * $16 = 2^2 + 4\cdot 2$
* Skaičiavimai. $2^2 + 4\cdot 2 = 4 + 8 = 12$, todėl gauta lygybė $16 = 2^2 + 4\cdot 2$ $\color{red}{\text{neteisinga}}$
* Išvada pagal **pagalbinį teiginį**. Taškas $(2;\,16)$ $\color{red}{\text{nepriklauso}}$ funkcijos $y = x^2 + 4x$ grafikui.

**Paaiškinimas 1C**
Kaip patikrinti, ar taškas $(4;\,24)$ priklauso funkcijos $y = x^2 + 4x$ grafikui?
* Pagal **pagalbinį teiginį** formulėje pakeisti: $\begin{cases} {\color{blue}{x}}\to 4 \\ {\color{green}{y}}\to 24\end{cases}$
* Pakeista funkcijos formulė:
    * $\boxed{24} = \boxed{4}^2 + 4\boxed{4}$
    * $24 = 4^2 + 4\cdot 4$
* Skaičiavimai. $4^2 + 4\cdot 4 = 16 + 16 = 32$, todėl gauta lygybė $24 = 4^2 + 4\cdot 4$ $\color{red}{\text{neteisinga}}$
* Išvada pagal **pagalbinį teiginį**. Taškas $(4;\,24)$ $\color{red}{\text{nepriklauso}}$ funkcijos $y = x^2 + 4x$ grafikui.

**Paaiškinimas 1D**
Kaip patikrinti, ar taškas $(3;\,18)$ priklauso funkcijos $y = x^2 + 4x$ grafikui?
* Pagal **pagalbinį teiginį** formulėje pakeisti: $\begin{cases} {\color{blue}{x}}\to 3 \\ {\color{green}{y}}\to 18\end{cases}$
* Pakeista funkcijos formulė:
    * $\boxed{18} = \boxed{3}^2 + 4\boxed{3}$
    * $18 = 3^2 + 4\cdot 3$
* Skaičiavimai. $3^2 + 4\cdot 3 = 9 + 12 = 21$, todėl gauta lygybė $18 = 3^2 + 4\cdot 3$ $\color{red}{\text{neteisinga}}$
* Išvada pagal **pagalbinį teiginį**. Taškas $(3;\,18)$ $\color{red}{\text{nepriklauso}}$ funkcijos $y = x^2 + 4x$ grafikui.

### Uždavinys 4. 
4\. Duota kvadratinė lygtis $x^2 + 3x - 4 = 0$.
* Naudodamiesi kvadratinės lygties diskriminanto formule, parodykite, kad šios lygties diskriminantas $D$ lygus $25$. (**1 tšk.** su sprendimu)
* Naudodamiesi kvadratinės lygties sprendinių formulėmis, apskaičiuokite šios lygties sprendinius.  (**2 tšk.** su sprendimu)

**Turinio žinios:**  
* **Diskriminanto formulė.** $D = b^2 - 4ac$
* **Kvadratinių sprendinių formulės.** $x_1 = \dfrac{-b - \sqrt{D}}{2a}$, $x_2 = \dfrac{-b + \sqrt{D}}{2a}$
 
**Sprendimo planas:**  
* Pritaikius **diskriminanto formulę** gauti, kad $D=25$
* Su gauta $D$ reikšme taikyti **kvadratinių sprendinių formules**

**Sprendimas**
* $D=3^2-4\cdot 1\cdot (-4) = 9 - (-16) = 9 + 16 = 25$  
* $x_1 = \dfrac{-3 - \sqrt{25}}{2} = \dfrac{-3 - 5}{2} = \dfrac{-8}{2} = -4$,
* $x_2 = \dfrac{-3 + \sqrt{25}}{2} = \dfrac{-3 + 5}{2} = \dfrac{2}{2} = 1$

**Paaiškinimas 1.**
Kaip rasti lygties $x^2 + 3x - 4 = 0$ diskriminantą?

* Lygties ${\color{blue}{a}}x^2+ {\color{green}{b}}x + {\color{red}{c}}=0$ diskriminanto formulė: $D={\color{green}{b}}^2-4{\color{blue}{a}}{\color{red}{c}}$
* $\boxed{\begin{array}{l}{\color{blue}{a}}x^2+ {\color{green}{b}}x + {\color{red}{c}}\\ x^2+3x-4\end{array}}\Leftrightarrow \boxed{\begin{array}{l}{\color{blue}{a}}x^2+ {\color{green}{b}}x + {\color{red}{c}}\\ {\color{blue}{1}}x^2+{\color{green}{3}}x+{\color{red}{(-4)}}\end{array}}$ , todėl formulėje pakeisti: $\begin{cases} {\color{blue}{a}}\to 1 \\ {\color{green}{b}}\to 3 \\ {\color{red}{c}}\to -4\end{cases}$
* Pakeista formulė:
    * $D=\boxed{3}^2-4\boxed{1}\boxed{-4}$
    * $D=3^2-4\cdot 1\cdot (-4)$
* Skaičiavimai:
    * $D=3^2-4\cdot 1\cdot (-4) = 9 - (-16) = 9 + 16 = 25$
 
**Paaiškinimas 2.**
Kaip rasti lygties $x^2 + 3x - 4 = 0$ sprendinius žinant, kad diskriminantas $D$ lygus $25$?
* Lygties ${\color{blue}{a}}x^2+ {\color{green}{b}}x + {\color{red}{c}}=0$ sprendinių formulės:
    * $x_1 = \dfrac{-{\color{green}{b}} - \sqrt{{\color{purple}{D}}}}{2{\color{blue}{a}}}$
    * $x_2 = \dfrac{-{\color{green}{b}} + \sqrt{{\color{purple}{D}}}}{2{\color{blue}{a}}}$
* Pakeisti: $\begin{cases} {\color{blue}{a}}\to 1 \\ {\color{green}{b}}\to 3 \\ {\color{red}{c}}\to -4 \\ {\color{purple}{D}}\to 25\end{cases}$
* Pakeistos formulės:
  * Sprendiniui $x_1$:
    * $x_1 = \dfrac{-\boxed{3} - \sqrt{\boxed{25}}}{2\boxed{1}}$
    * $x_1 = \dfrac{3 - \sqrt{25}}{2}$
  * Sprendiniui $x_2$:
    * $x_2 = \dfrac{-\boxed{3} + \sqrt{\boxed{25}}}{2\boxed{1}}$
    * $x_2 = \dfrac{-3 + \sqrt{25}}{2}$
* Skaičiavimai:
    * $x_1 = \dfrac{-3 - \sqrt{25}}{2} = \dfrac{-3 - 5}{2} = \dfrac{-8}{2} = -4$,
    * $x_2 = \dfrac{-3 + \sqrt{25}}{2} = \dfrac{-3 + 5}{2} = \dfrac{2}{2} = 1$

### Uždavinys 18. 
Paveiksle pavaizduotas parabolės $y=ax^2+c$ eskizas. 

<img src="images/2025_18_1.png" width="150"/>
    
Pažymėkite paveikslą, kuriame pavaizduotas parabolės $y=cx^2+a$ eskizas.

<img src="images/2025_18_2.png" width="600"/>

**Turinio žinios**
* **Teiginys 1.** Formos $y=ax^2+bx+c$ parabolėje koeficientas $a$ parodo, kur bus nukreiptos parabolės eskizo šakos. Jei $a>0$, šakos nukreiptos aukštyn. Jei $a<0$, šakos nukreiptos žemyn.
* **Teiginys 2.** Formos $y=ax^2+c$ parabolėje koeficientas $c$ parodo parabolės eskizo viršūnės padėtį. Jei $c>0$, parabolės eskizo viršūnė yra virš taško $(0; 0)$. Jei $c<0$, tai po tašku $(0; 0)$

**Teiginių paaiškinimas**
* Nuorodoje [https://www.desmos.com/calculator/cyra0x3j1m](https://www.desmos.com/calculator/cyra0x3j1m) išsiaiškinti, kaip keičiant parabolės parametrus $a$ ir $c$ transformuojasi jos eskizas.
  
**Sprendimo planas**  
* Remiantis abiem teiginiais nustatyti $a$ ir $c$ ženklus parabolėje $y=ax^2+c$
* Radus $a$ ir $c$ ženklus, nustatyti parabolės $y=cx^2+a$ eskizo elementus: viršūnės ir šakų padėtį.

**Sprendimas**
* Parabolės $y=ax^2+c$ eskize šakos nukreiptos aukštyn, todėl $a>0$
* Parabolės $y=ax^2+c$ eskize viršūnė yra po tašku $(0; 0)$, todėl $c<0$
* Naujoje parabolėje $y=cx^2+a$ koeficientas prie $x^2$ lygus $c$ ir mažesnis už $0$, todėl jos eskize šakos nukreiptos žemyn.
* Naujoje parabolėje $y=cx^2+a$ laisvasis narys lygus $a$ ir didesnis už $0$, todėl jos eskize viršūnė yra virš taško $(0; 0)$

### Uždavinys 21. 
Paveiksle pavaizduoto skritulio centras yra taškas $O$, o spindulio ilgis $12$. Atkarpa $AB$ yra šio skritulio styga, $\angle AOB = 60^\circ$. Geltonai nuspalvintos nuopjovos plotas užrašytas pavidalu $a \cdot \pi - b \cdot \sqrt{3}$
 (čia $a$ ir $b$ - natūralieji skaičiai). Nustatykite $a$ ir $b$ reikšmes.

<img src="images/2025_21.png" width="180"/>

**Turinio žinios**
* **Trikampio ploto formulė.** $S = \frac{1}{2} ab \cdot \sin \angle C$
* **Skritulio išpjovos formulė.** $S_{išpj} = \dfrac{\pi R^2}{360} \cdot \alpha$

**Papildomos žinios**
* **Lygiakraščio trikampio ploto formulė.** $S = \dfrac{a^2\sqrt{3}}{4}$
* **Lygiašonio trikampio savybė**. Lygiašonio trikampio kampai prie pagrindo lygūs.
* **Lygiakraščio trikampio savybė**. Lygiakraščio trikampio visi kampai lygūs.
* **Teiginys**. Trikampio kampų suma lygi $180^\circ$
  
**Sprendimo planas**
* Rasti išpjovos plotą
* Rasti trikampio plotą
* Atėmus gautus plotus rasti geltonai nuspalvintos nuopjovos plotą

**Sprendimas 1**
* $S_{\triangle ABO} = \dfrac{1}{2} ab \cdot \sin \angle C = \dfrac{1}{2} \cdot 12 \cdot 12 \cdot \sin \angle 60^\circ = \cancel{6}^{\color{blue}{3}} \cdot 12 \cdot \dfrac{\sqrt{3}}{\cancel{2}^{\color{blue}{1}}} = 36\sqrt{3}$
* $S_{išpj} = \dfrac{\pi R^2}{360} \cdot \alpha = \dfrac{\pi\cdot 12^2}{\cancel{360}^{\color{blue}{6}}} \cdot \cancel{60}^{\color{blue}{1}} = \dfrac{\pi\cdot 12\cdot \cancel{12}^{\color{blue}{2}}}{\cancel{6}^{\color{blue}{1}}} = \pi\cdot 12 \cdot 2 = 24\pi$
* $S_{nuopj} = 24\pi - 36\sqrt{3}$
* $a=24$; $b=36$

**Sprendimas 2**  
Panašus, tik kitokia trikampio ploto skaičiavimo eiga
* $\triangle ABO$ - lygiašonis, todėl $\angle OAB = \angle OBA$
* $\angle ABO + \angle OAB = 180^\circ - 60^\circ = 120^\circ$
* $\angle OAB = \angle OBA = \dfrac{120^\circ}{2} = 60^\circ$
* $\triangle ABO$ - lygiakraštis, nes visi jo kampai lygūs
* $S_{\triangle ABO} = \dfrac{a^2\sqrt{3}}{4} = \dfrac{12^2\sqrt{3}}{4} = \dfrac{\cancel{12}^{\color{blue}{3}} \cdot 12 \cdot \sqrt{3}}{\cancel{4}^{\color{blue}{1}}} = 36\sqrt{3}$
