### Uždavinys 1.
Duota funkcija $y = f(x) = x^2 + 4x$. Kuris iš nurodytų taškų priklauso šios funkcijos grafikui? (**1 tšk.**)  
$\begin{array}{l{\quad}l{\quad}l{\quad}l}
\boxed{A}\, (1;\,5) &
\boxed{B}\, (2;\,16) & 
\boxed{C}\, (4;\,24) &
\boxed{D}\, (3;\,18)
\end{array}$ 

* **Sprendimo planas:**
    * **Pagalbinis teiginys**. Taškas $({\color{blue}{a}}; {\color{green}{b}})$ priklauso pateiktos funkcijos $y = f(x)$ grafikui tada ir tik tada, kai pakeitus $\begin{cases} {\color{blue}{x}}\to a \\ {\color{green}{y}}\to b \end{cases}$ lygybė tampa teisinga.
    * Patikrinti, kuriam iš pateiktų variantų lygybė yra teisinga.
      
* **Sprendimas.**
    * $5 = 1^2 + 4\cdot 1$, todėl $(1;\,5)$ priklauso funkcijos $y = x^2 + 4x$ grafikui.
 
* **Paaiškinimas.**
Kaip patikrinti, ar taškai $(1;\,5)$, $(2;\,16)$, $(4;\,24)$, $(3;\,18)$ priklauso funkcijos grafikui?
    * Funkcijos formulė: $\boxed{{\color{green}{y}} = {\color{blue}{x}}^2 + 4{\color{blue}{x}}}$
    * Atsakymo variantų patikrinimai:
        * Taškui $(1;\,5)$:
            * Pagal **pagalbinį teiginį** formulėje pakeisti: $\begin{cases} {\color{blue}{x}}\to 1 \\ {\color{green}{y}}\to 5\end{cases}$
            * Pakeista funkcijos formulė:
                * $\boxed{5} = \boxed{1}^2 + 4\boxed{1}$
                * $5 = 1^2 + 4\cdot 1$
            * Skaičiavimai. $1^2 + 4\cdot 1 = 1 + 4 = 5$, todėl gauta lygybė $5 = 1^2 + 4\cdot 1$ $\color{green}{\text{teisinga}}$
            * Išvada pagal **pagalbinį teiginį**. Taškas $(1;\,5)$ $\color{green}{\text{priklauso}}$ funkcijos $y = x^2 + 4x$ grafikui.
        * Taškui $(2;\,16)$:
            * Pagal **pagalbinį teiginį** formulėje pakeisti: $\begin{cases} {\color{blue}{x}}\to 2 \\ {\color{green}{y}}\to 16\end{cases}$
            * Pakeista funkcijos formulė:
                * $\boxed{16} = \boxed{2}^2 + 4\boxed{2}$
                * $16 = 2^2 + 4\cdot 2$
            * Skaičiavimai. $2^2 + 4\cdot 2 = 4 + 8 = 12$, todėl gauta lygybė $16 = 2^2 + 4\cdot 2$ $\color{red}{\text{neteisinga}}$
            * Išvada pagal **pagalbinį teiginį**. Taškas $(2;\,16)$ $\color{red}{\text{nepriklauso}}$ funkcijos $y = x^2 + 4x$ grafikui.
        * Taškui $(4;\,24)$:
            * Pagal **pagalbinį teiginį** formulėje pakeisti: $\begin{cases} {\color{blue}{x}}\to 4 \\ {\color{green}{y}}\to 24\end{cases}$
            * Pakeista funkcijos formulė:
                * $\boxed{24} = \boxed{4}^2 + 4\boxed{4}$
                * $24 = 4^2 + 4\cdot 4$
            * Skaičiavimai. $4^2 + 4\cdot 4 = 16 + 16 = 32$, todėl gauta lygybė $24 = 4^2 + 4\cdot 4$ $\color{red}{\text{neteisinga}}$
            * Išvada pagal **pagalbinį teiginį**. Taškas $(4;\,24)$ $\color{red}{\text{nepriklauso}}$ funkcijos $y = x^2 + 4x$ grafikui.
        * Taškui $(3;\,18)$:
            * Pagal **pagalbinį teiginį** formulėje pakeisti: $\begin{cases} {\color{blue}{x}}\to 3 \\ {\color{green}{y}}\to 18\end{cases}$
            * Pakeista funkcijos formulė:
                * $\boxed{18} = \boxed{3}^2 + 4\boxed{3}$
                * $18 = 3^2 + 4\cdot 3$
            * Skaičiavimai. $3^2 + 4\cdot 3 = 9 + 12 = 21$, todėl gauta lygybė $18 = 3^2 + 4\cdot 3$ $\color{red}{\text{neteisinga}}$
            * Išvada pagal **pagalbinį teiginį**. Taškas $(3;\,18)$ $\color{red}{\text{nepriklauso}}$ funkcijos $y = x^2 + 4x$ grafikui.

$\begin{array}{|c|c|c|c|} \hline x & -2 & 0 & 4 \\ \hline f(x) = 2x + 5 & 1 & 5 & \\ \hline\end{array}$  
Į lentelės tuščią langelį įrašykite trūkstamą skaičių (**1 tšk.**)

    * $f(4) = 2\cdot 4 + 5 = 13$

    * Pakeisti: ${\color{blue}{x}}\to 4$

4\. Duota kvadratinė lygtis $x^2 + 3x - 4 = 0$.
* Naudodamiesi kvadratinės lygties diskriminanto formule, parodykite, kad šios lygties diskriminantas $D$ lygus $25$. (**1 tšk.** su sprendimu)
* Naudodamiesi kvadratinės lygties sprendinių formulėmis, apskaičiuokite šios lygties sprendinius.  (**2 tšk.** su sprendimu)

**Sprendimo planas:**  
* Taikyti **diskriminanto formulę** $D = b^2 - 4ac$
* Taikyti **kvadratinių sprendinių formules** $x_1 = \dfrac{-b - \sqrt{D}}{2a}$, $x_2 = \dfrac{-b + \sqrt{D}}{2a}$

**Sprendimas**
* $D=3^2-4\cdot 1\cdot (-4) = 9 - (-16) = 9 + 16 = 25$  
* $x_1 = \dfrac{-3 - \sqrt{25}}{2} = \dfrac{-3 - 5}{2} = \dfrac{-8}{2} = -4$,
* $x_2 = \dfrac{-3 + \sqrt{25}}{2} = \dfrac{-3 + 5}{2} = \dfrac{2}{2} = 1$

**Paaiškinimas 1.**
Kaip rasti lygties $x^2 + 3x - 4 = 0$ diskriminantą?

* Lygties ${\color{blue}{a}}x^2+ {\color{green}{b}}x + {\color{red}{c}}=0$ diskriminanto formulė: $D={\color{green}{b}}^2-4{\color{blue}{a}}{\color{red}{c}}$
* $\boxed{\begin{array}{l}{\color{blue}{a}}x^2+ {\color{green}{b}}x + {\color{red}{c}}\\ x^2+3x-4\end{array}}\Leftrightarrow \boxed{\begin{array}{l}{\color{blue}{a}}x^2+ {\color{green}{b}}x + {\color{red}{c}}\\ {\color{blue}{1}}x^2+{\color{green}{3}}x+{\color{red}{(-4)}}\end{array}}$ , todėl formulėje pakeisti: $\begin{cases} {\color{blue}{a}}\to 1 \\ {\color{green}{b}}\to 3 \\ {\color{red}{c}}\to -4\end{cases}$
* Pakeista formulė:
    * $D=\boxed{3}^2-4\boxed{1}\boxed{-4}$
    * $D=3^2-4\cdot 1\cdot (-4)$
* Skaičiavimai:
    * $D=3^2-4\cdot 1\cdot (-4) = 9 - (-16) = 9 + 16 = 25$
 
**Paaiškinimas 2.**
Kaip rasti lygties $x^2 + 3x - 4 = 0$ sprendinius žinant, kad diskriminantas $D$ lygus $25$?
* Lygties ${\color{blue}{a}}x^2+ {\color{green}{b}}x + {\color{red}{c}}=0$ sprendinių formulės:
    * $x_1 = \dfrac{-{\color{green}{b}} - \sqrt{{\color{purple}{D}}}}{2{\color{blue}{a}}}$
    * $x_2 = \dfrac{-{\color{green}{b}} + \sqrt{{\color{purple}{D}}}}{2{\color{blue}{a}}}$
* Pakeisti: $\begin{cases} {\color{blue}{a}}\to 1 \\ {\color{green}{b}}\to 3 \\ {\color{red}{c}}\to -4 \\ {\color{purple}{D}}\to 25\end{cases}$
* Pakeistos formulės:
  * Sprendiniui $x_1$:
    * $x_1 = \dfrac{-\boxed{3} - \sqrt{\boxed{25}}}{2\boxed{1}}$
    * $x_1 = \dfrac{3 - \sqrt{25}}{2}$
  * Sprendiniui $x_2$:
    * $x_2 = \dfrac{-\boxed{3} + \sqrt{\boxed{25}}}{2\boxed{1}}$
    * $x_2 = \dfrac{-3 + \sqrt{25}}{2}$
* Skaičiavimai:
    * $x_1 = \dfrac{-3 - \sqrt{25}}{2} = \dfrac{-3 - 5}{2} = \dfrac{-8}{2} = -4$,
    * $x_2 = \dfrac{-3 + \sqrt{25}}{2} = \dfrac{-3 + 5}{2} = \dfrac{2}{2} = 1$
