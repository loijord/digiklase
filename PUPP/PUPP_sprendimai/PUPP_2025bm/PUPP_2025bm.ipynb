### Uždavinys 1.
Mokykloje mokosi $128$ mergaitės, kurios sudaro $80\%$ berniukų skaičiaus. Kiek moksleivių mokosi mokykloje? (**1 tšk.**) 

* **Sprendimo planas:**
    * $\overset{0.8x}{\overbrace{\text{mergaitės}}}$ $\color{blue}{\underline{\text{sudaro}\,80\%}}$ $\overset{x}{\overbrace{\text{berniukų skaičiaus}}}$
    * Lygtyje mergaičių skaičių prilyginti 128.
    * $\text{Moksleiviai} = \text{mergaitės} + \text{berniukai}$
      
* **Sprendimas.**
    * $\left\{\begin{array}{rl} 
    0.8x & \text{mergaitės} \\ 
    x & \text{berniukai} \\ 
    \end{array}\right.$
    
    * $\boxed{\begin{array}{rcl}
    0.8x &=& 128|:0.8 \\
    x &=& 128 : 0.8 \\
    x &=& 160
    \end{array}}$
    
    * $\left\{\begin{array}{rl} 
    128 &\text{mergaitės} \\ 
    160 &\text{berniukai} \\ 
    \end{array}\right.$

    * $\color{blue}{\text{Kiek moksleivių}}$: $128 + 160 = 288$

### Uždavinys 2.
Suprastink reiškinį $\dfrac{x^2-4x}{x^2-16}$ (**1 tšk.**) 

* **Sprendimo planas:**
    * Išskaidyti $x^2-4x$ dauginamaisiais
    * Išskaidyti $x^2-16$ dauginamaisiais
    * Suprastinti gautą trupmeną

* **Sprendimas**  
$\dfrac{x^2-4x}{x^2-16} = \dfrac{x(x-4)}{(x-4)(x+4)} = \dfrac{x}{x+4}$

* **Paaiškinimas 1.**
Kaip išskaidyti $x^2 - 16$ dauginamaisiais?

    * Formulė: $a^2 - b^2 = (a - b)(a + b)$
    * $\boxed{\begin{array}{l}{\color{blue}{a}}^2 - b^2 \\ {\color{blue}{x}}^2-16\end{array}} \Leftrightarrow \boxed{\begin{array}{l} {\color{blue}{a}}^2 - {\color{green}{b}}^2 \\ {\color{blue}{x}}^2-{\color{green}{4}}^2 \end{array}}$, todėl formulėje pakeisti: $\begin{cases} {\color{blue}{a\to x}} \\ {\color{green}{b\to 4}}\end{cases}$
    * Pakeista formulė:
        * $x^2 - 4^2 = (x - 4)(x + 4)$
     
* **Paaiškinimas 2.**
Kaip išskaidyti $x^2 - 4x$ dauginamaisiais?

    * Formulė: $ab - ac = a(b-c)$
    * $\boxed{\begin{array}{l} ab - ac \\ x^2-4x \end{array}} \Leftrightarrow
     \boxed{\begin{array}{l} {\color{blue}{a}}\cdot {\color{orange}{b}} - {\color{blue}{a}} \cdot {\color{red}{c}} \\ {\color{blue}{x}}\cdot {\color{orange}{x}} - {\color{blue}{x}}\cdot {\color{red}{4}}\end{array}}$, todėl formulėje pakeisti: $\begin{cases} {\color{blue}{a\to x}} \\ {\color{orange}{b\to x}} \\ {\color{red}{c\to 4}}\end{cases}$
    * Pakeista formulė:
        * $\boxed{x}\boxed{x} - \boxed{x}\boxed{4} = \boxed{x}(\boxed{x}-\boxed{4})$
        * $x^2 - 4x = x(x-4)$

### Uždavinys 4.
Apskaičiuokite lygties $(x+1)(x-7)=5$ diskriminanto reikšmę. (**1 tšk.**) 

**Sprendimo planas:**  
* Suvesti lygtį į kanoninį pavidalą $ax^2+bx+c=0$
* Taikyti diskriminanto formulę

**Sprendimas**
* $\boxed{\begin{array}{rcl}
    (x+1)(x-7) &=& 5 \\
    x^2 - 6x -7 &=& 5|\tiny{-5} \\
    x^2 - 6x -7-5 &=& 0 \\
    x^2 - 6x -12 &=& 0
    \end{array}}$  
* $D=(-6)^2-4\cdot 1\cdot (-12) = 84$  

**Paaiškinimas 1.**
Kaip atskliausti $(x+1)(x-7)$?

* $(x+1)(x-7) = \begin{array}{r||c|c}
&x&-7\\ \hline
x&x^2&-7x\\ \hline
1&x&-7
\end{array} = x^2 -7x + x - 7 = x^2 - 6x -7$

**Paaiškinimas 2.**
Kaip rasti lygties $x^2 - 6x -7 = 0$ diskriminantą?

* Lygties $ax^2+bx+c=0$ diskriminanto formulė: $D=b^2-4ac$
* $\boxed{\begin{array}{l}{\color{blue}{a}}x^2+ {\color{green}{b}}x + {\color{red}{c}}\\ x^2-6x-12\end{array}}\Leftrightarrow \boxed{\begin{array}{l}{\color{blue}{a}}x^2+ {\color{green}{b}}x + {\color{red}{c}}\\ {\color{blue}{1}}x^2+{\color{green}{(-6)}}x+{\color{red}{(-12)}}\end{array}}$ , todėl formulėje pakeisti: $\begin{cases} {\color{blue}{a\to 1}} \\ {\color{green}{b\to -6}} \\ {\color{red}{c\to -12}}\end{cases}$
* Nauja formulė:
    * $D=\boxed{-6}^2-4\boxed{1}\boxed{-12}$
    * $D=(-6)^2-4\cdot 1\cdot (-12) = 84$

**Sprendimo planas:**  
* Išvesti $x$ iš pateiktos sąlygos.
* Įstatyti į duotą reiškinį ir apskaičiuoti naujo reiškinio reikšmę.

**Sprendimas**

* $\boxed{\begin{array}{rcl}
    0{,}2x &=& 2y|\tiny{:0{,}2} \\
    x &=& 2y:0{,}2 \\
    x &=& 2y\cdot 5\\
    x &=& 10y\end{array}
}$
* $\dfrac{5x-4y}{2x+3y} = \dfrac{5\cdot 10y-4y}{2\cdot 10y+3y} = \dfrac{50y-4y}{20y+3y} = \dfrac{46y}{23y} = 2$

**Paaiškinimas**
Kaip reiškinyje $\dfrac{5x-4y}{2x+3y}$ įstatyti $x\to 10y$?

* Reiškinys: $\dfrac{5x-4y}{2x+3y}$
* Pakeisti: $x\to 10y$
* Naujas reiškinys:
  * $\dfrac{5\boxed{10y}-4y}{2\boxed{10y}+3y}$
  * $\dfrac{5\cdot 10y-4y}{2\cdot 10y+3y}$

### Uždavinys 6.
Kiekvienas sekos $x$, $y$, $z$, $-1$, $8$, $-9$, $17$, $\dots$ narys, pradedant trečiuoju, lygus dviejų prieš jį einančių narių skirtumui. Kam lygus skaičius $x$? (**1 tšk.**)  

**Spendimo planas**
* $\overset{a_n}{\overbrace{\text{Kiekvienas narys}}}$ $\color{blue}{\underline{\text{lygus}}}$ $\overset{a_{n-2}-a_{n-1}}{\overbrace{\text{dviejų prieš jį einančių narių skirtumui}}}$
* Užrašyti 3, 4 ir 5 nario sąryšį, iš jo rasti 3 narį.
* Užrašyti 2, 3 ir 4 nario sąryšį, iš jo rasti 2 narį.
* Užrašyti 1, 2 ir 3 nario sąryšį, iš jo rasti 1 narį.

**Spendimas**
* Penktam nariui:  
  $\boxed{\begin{array}{rcl} z -(-1) &=& 8 \\ z + 1 &=& 8|\tiny{-1} \\ z &=& 8 - 1 \\ z &=& 7 \end{array}}$
* Ketvirtam nariui:  
  $\boxed{\begin{array}{rcl} y - 7 &=& -1|\tiny{+7} \\ y &=& -1 + 7 \\ y &=& 6 \end{array}}$
* Trečiam nariui:  
  $\boxed{\begin{array}{rcl} x - 6 &=& 7|\tiny{-6} \\ x &=& 7 + 6 \\ x &=& 13 \end{array}}$

### Uždavinys 7.
Su kuria parametro $a$ reikšme lygtis $2x^2+2ax+3a=0$ turi vieną sprendinį? (**1 tšk.**)

**Sprendimo planas:**  
* Taikyti diskriminanto $D$ formulę
* Rasti (pateikto) $a$ reikšmes, su kuriomis $D=0$

**Sprendimas**
* $D=(2a)^2-4\cdot 2\cdot 3a = 4a^2 - 24a$
* $\boxed{\begin{array}{rcl}
    4a^2 - 24a &=& 0|:4\\
    a^2 - 6a &=& 0 \\
    a(a-6) &=& 0 \\
    \boxed{\begin{array}{rcl} a&=&0\end{array}} &\text{arba} & \boxed{\begin{array}{rcl} a-6&=&0|\tiny{+6}\\a&=&6\end{array}}
    \end{array}}$  

**Paaiškinimas 1.**
Kaip rasti lygties $2x^2+2ax+3a=0$ diskriminantą?

* Lygties $ax^2+bx+c=0$ diskriminanto formulė: $D=b^2-4ac$
* $\boxed{\begin{array}{l}{\color{blue}{a}}x^2+ {\color{green}{b}}x + {\color{red}{c}}\\ {\color{blue}{2}}x^2+{\color{green}{2a}}x+{\color{red}{3a}}=0\end{array}}$ , todėl formulėje pakeisti: $\begin{cases} {\color{blue}{a\to 2}} \\ {\color{green}{b\to 2a}} \\ {\color{red}{c\to 3a}}\end{cases}$
* Pakeista diskriminanto formulė:
    * $D=\boxed{2a}^2-4\boxed{2}\boxed{3a}$
    * $D=(2a)^2-4\cdot 2\cdot 3a$
      
**Paaiškinimas 2.**
Kaip išskaidyti $a^2 - 6a$ dauginamaisiais?

* Formulė: $ab - ac = a(b-c)$
* $\boxed{\begin{array}{l} ab - ac \\ a^2-6a \end{array}} \Leftrightarrow
 \boxed{\begin{array}{l} {\color{blue}{a}}\cdot {\color{orange}{b}} - {\color{blue}{a}} \cdot {\color{red}{c}} \\ {\color{blue}{a}}\cdot {\color{orange}{a}} - {\color{blue}{a}}\cdot {\color{red}{6}}\end{array}}$, todėl formulėje pakeisti: $\begin{cases} {\color{blue}{a\to a}} \\ {\color{orange}{b\to a}} \\ {\color{red}{c\to 6}}\end{cases}$
* Pakeista formulė:
    * $\boxed{a}\boxed{a} - \boxed{a}\boxed{6} = \boxed{a}(\boxed{a}-\boxed{6})$
    * $a^2 - 6a = a(a-6)$

### Uždavinys 9.  
Suprastinkite reiškinį $(\sin x + \cos x)^2 - 2\text{tg}x\cdot \cos ^2{x}$  (**1 tšk.**) 

**Sprendimo planas:** 
* Taikyti tangento apibrėžimą: $\text{tg} = \dfrac{\sin x}{\cos x}$
* Taikyti greitosios daugybos formulę $(a+b)^2 = a^2+2ab+b^2$ reiškiniui $(\sin x + \cos x)^2$
* Taikyti formulę $\sin^2 x + \cos^2 x = 1$

**Sprendimas.**  
$\begin{array}{lcr}(\sin x + \cos x)^2 - 2\text{tg}x\cdot \cos ^2{x} =&& \\ 
(\sin x + \cos x)^2 - 2\dfrac{\sin x}{\cos x}\cdot \cos ^2{x} =&& \\
\sin^2 x + 2\sin x\cos x + \cos^2 x - 2\sin x \cos x =&& \\
\sin^2 x + \cos^2 x = 1 &&
\end{array}$
