# Užduotis
Žvejys sugavo ir pardavė $N$ žuvų, kurių kiekvienos kainą nustatė tokiu būdu. Pirma žuvis įvertinta $a_0 = M\,\text{ct}$. Kiekvienos kitos žuvies kaina $a_i$ yra įvertinama pagal rekurentinį sąryšį $a_i = a_{i-1} \times (i+1),\,i \in \mathbb{N}$. Raskite visų sugautų ir parduotų žuvų bendrą kainą centais.

### Sprendimas 1
Kiekvienai žuviai nustatome jos vertę ir gautą skaičių pridedame prie bendros sumos. Algoritmas yra taupus atminties prasme, nes reikia saugoti tik du skaičius: bendrą sumą ir $i$ - tosios žuvies kainą.

```python
N = 7 #kiek žuvų pagavo per visą gaudymą
M = 2 #atlyginimas ct už pirmą žuvį

m = M #kintanti žuvies kaina
su = m #reikiama suma

for i in range(1, N):
    m = m * (i + 1)
    su += m #su = su + m

print(f'Suma = {su} ct')
```

### Sprendimas 2
Sudarome skaičių seką, kurioje saugomos žuvų kainos. Tada suskaičiuojame tos sekos narių bendrą sumą.

```python
N = 7 #kiek žuvų pagavo per visą gaudymą
M = 2 #atlyginimas ct už pirmą žuvį

a = [M] + [0] * (N - 1) #kintanti žuvies kaina
print(f'pradžioje:', a)

for i in range(1, N):
    a[i] = a[i-1] * (i + 1)

print(f'paskui:', a)
print(f'Suma = {sum(a)} ct')
```
