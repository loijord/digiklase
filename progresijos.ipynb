### Aritmetinė progresija

* Visų aritmetinių progresijų ,,protėvis": $0;\,1;\,2;\,\dots;\,n$
* Pagrindiniai parametrai:  
    * Pirmas narys: $a_1=0$
    * Skirtumas: $d = 1$
    * Narių skaičius: $n+1$
    * Narių suma: $n \cdot \dfrac{n+1}{2}$
* Pagrindinė savybė, naudojama narių sumos formulės įrodyme: $i$-tojo ir $i$-tojo nuo galo narių vidurkis yra pastovus skaičius, nekintantis su kiekvienu natūraliuoju $\left(i \in [1; n]\right)$. Šį skaičių patogiausia suskaičiuoti pagal pirmo ir paskutinio nario vidurkį.


**Įrodymas (negriežtas)**
  * Narių skaičius: $n+1$
  * Pirmo ir paskutinio narių vidurkis: $\dfrac{n}{2}$.
  * Jei skaičių rinkinį galima suskirstyti į grupes taip, kad kiekvienos grupės vidurkis būtų vienodas, tai tam vidurkiui lygus ir viso rinkinio narių vidurkis
  * $\text{Vidurkis} = \dfrac{\text{Suma}}{\text{Narių skaičius}}$, todėl norint gauti sumą, reikia padauginti narių skaičių $n+1$ iš vidurkio $\dfrac{n}{2}$

**Įrodymas (griežtas)**  
* $\begin{array}{rcl} 
2 \cdot (0 + 1 + \dots + n) &=& (0 + 1 + \dots + n) + ((n-0) + (n-1) + \dots + (n-n)) = \\
 &=&\left(0 + n - 0\right) + \left(1 + n - 1\right) + \dots \left(n + n - n\right) = \\
 &=& \underbrace{n + n + \dots + n}_{n+1} = (n+1) \cdot n
\end{array}$
* $\displaystyle 2 \sum_{i=0}^{n} i = \sum_{i=0}^{n} i + \sum_{i=0}^{n}\left(n - i\right) = \sum_{i=0}^{n} \left(i + n - i\right) = \sum_{i=0}^{n} n = (n+1) \cdot n$

**Kodėl patogiau ,,žaisti" su 0 pirmajame naryje?**  
* Lengvesnis griežtas įrodymas
* Bet daug patogiau žinoti, kaip susumuoti $n$ pirmų narių vietoj $n+1$

**Likusių aritmetinių progresijų savybės**  
* Nagrinėsime seką su $n$ narių: $0; 1; 2; \dots; n-1$, jos narių suma bus $\dfrac{n(n-1)}{2}$.
* Norint ,,praretinti" / ,,sutankinti" narius užtenka padauginti visus narius iš norimo $d$.
  * $0; 1; 2; \dots; n$ pasikeičia į $0; d; 2d; \dots; (n-1)d$
  * Suma irgi pasidaugina iš $d$: $0 + d + 2d + \dots + (n-1)d = n \cdot \dfrac{(n-1)d}{2}$
* Norint ,,paslinkti" narius užtenka prie gauto sekos varianto narių pridėti po $a_1$:
  * $0; d; 2d; \dots; (n-1)d$ pasikeičia į $a_1; a_1+d; a_1+2d; \dots; a_1+(n - 1)d$
  * Prie gautos sumos prisideda $n$ kartų po $a_1$: $a_1 + (a_1+d) + (a_1+2d) + \dots + (a_1+(n - 1)d) = na_{1} + n \cdot \dfrac{(n-1)d}{2} = n \left(a_{1} + \dfrac{(n-1)d}{2}\right)$
  * Pirmo ir paskutinio narių vidurkis: $\dfrac{a_1 + (a_1+(n - 1)d)}{2} = \dfrac{2a_1 +(n - 1)d)}{2} = a_1 +\dfrac{(n - 1)d}{2}$

**Apibendrinimas**  
* Bet kuri aritmetinė progresija yra užrašoma forma $a_1; a_1+d; a_1+2d; \dots; a_1+(n - 1)d$.
* Ypač svarbu žinoti, kad $a_n = a_1+(n - 1)d$.
* Aritmetinės progresijos sumavimo lengviausias būdas: dauginti pirmo ir paskutinio narių vidurkį iš narių skaičiaus
* Bet kuri aritmetinė progresija su $n$ narių yra gaunama kiek reikia ,,praretinus" / ,,sutankinus", o paskui kiek reikia ,,paslinkus" sekos $0; 1; 2; \dots; n-1$ narius
* Sumos formulės žinoti nebūtina, bet būtina žinoti, kad $S_n = n \cdot \dfrac{a_1 + a_n}{2}$ 

### Geonetrinė progresija

* Visų aritmetinių progresijų ,,protėvis": $1;\,q;\,q^2;\,\dots;\,q^{n-1}$
* Pagrindiniai parametrai:  
    * Pirmas narys: $b_1=1$
    * Vardiklis: $q = 1$
    * Narių skaičius: $n$
    * Narių suma: $q^n - 1$
* Pagrindinė idėja, naudojama narių sumos formulės įrodyme: padauginti sumą $1 + q + q^2 + \dots + q^{n-1}$ iš $1 - q$


**Įrodymas 1**
  * $(1 + q + q^2 + \dots + q^{n-1})\cdot (1 - q) = 1 - q^n \Leftrightarrow 1 + q + q^2 + \dots + q^{n-1} = \dfrac{1-q^n}{1 - q}$

**Įrodymas 2**  
* Tegu $1 + q + q^2 + \dots + q^{n-1} = S$
* Pridėkime dar vieną narį $q^n$. Tada nauja suma bus $S + q^n$
* Galime kitą narį pridėti ir kitu būdu. Padauginame visus narius iš $q$ ir įtrepiame $1$ sekos pradžioje. Nauja suma bus $1 + qS$
* Išsprendę lygtį $S + q^n = 1 + qS$. Išsprendę šią lygtį nežinomojo $S$ atžvilgiu gausime $S = \dfrac{1-q^n}{1 - q}$

**Likusių geometrinių progresijų savybės**  
* Galima padauginti visus narius iš daugiklio $b_1$, tada nauja seka bus $b_1 + b_1q + b_1q^2 + \dots + b_1q^{n-1}$, o jos suma $S = b_1\cdot\dfrac{1-q^n}{1 - q}$
* Jei $|q| < 1$, tai didėjant $n$ nariai nyksta (artėja prie 0). Todėl tokia progresija vadinama nykstamąja.
  * Kai $n \to +\infty$, tai $q^n \to 0$ ir $S = b_1\cdot\dfrac{1-q^n}{1 - q} \to \dfrac{b_1}{1 - q}$

**Kitos pastabos**
* Jeigu didinant narių skaičių bet kurioje skaičių sekoje pamatome, kad jų suma artėja į tam tikrą skaičių (turi ribą), tai priimta laikyti, kad suma **lygi** tam skaičiui.
* Jeigu kalbame apie sudėtines palūkanas, vietoj $b_1$ imamas pradinis kapitalas $S$, o $q = 1 + \dfrac{p}{100}$, čia $p$ yra procentai.
* Kai $p>0$, kapitalas auga, kai $p<0$, nyksta.

**Apibendrinimas**  

* Bet kuri aritmetinė progresija yra užrašoma forma $b_1;\, b_1q;\, b_1q^2;\, \dots;\, b_1q^{n-1}$.
* Ypač svarbu žinoti, kad $b_n = b_1q^{n-1}$.
* Geometrinės progresijos $b_1 + b_1q + b_1q^2 + \dots + b_1q^{n-1}$ sumavimo lengviausias būdas: išvesti arba įsiminti formulę $S = \dfrac{1-q^n}{1 - q}$ sekai $1 + q + q^2 + \dots + q^{n-1}$, tada rezultatą dauginti iš $b_1$ ir gauti $S = b_1\cdot\dfrac{1-q^n}{1 - q}$
* Jei $|q| < 1$ vietoje $S = b_1\cdot\dfrac{1-q^n}{1 - q}$, geometrinę progresiją laikome nykstamąja ir imame $S = b_1\cdot\dfrac{1}{1 - q}$
